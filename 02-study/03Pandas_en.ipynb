## Practice Pandas

````python
import pandas as pd
# File url
url = 'https://github.com/E-man85/ML-PY/raw/main/03-data/sampledata.xls'
# Read directly from GitHub with pandas
df1 = pd.read_excel(url, sheet_name=0, engine='xlrd')
# Show first 5 rows of dataframe
df1.head()
````

````python
# rename columns
df1.rename(columns={'OrderDate':'Order_Date'}, inplace=True) #implace=true rewrites the original data frame
df1.rename(columns={'Unit Cost':'Unit_Cost'}, inplace=True)
df1.head()
````

````python
# First lines
df1.head()
````

````python
# last lines
df1.tail()
````

````python
# General information
df1.info()
````

````python
# Summary statistics
df1.describe()
````

````python
# summary statistics specifying variables
df1[['Units','Total']].describe()
````

````python
# change output formatting
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df1.head(5)
````

````python
# select lines
df1[0:3]
````

````python
# remove column
df1.drop(['Total'], axis = 1, inplace = True)# axis 1 means columns 0 means rows
df1.head(5)
````

````python
# operations between columns
df1['New_Total'] = df1['Units'] * df1['Unit_Cost'] # basic operation
df1['New_Total'].head()
````

````python
# select rows according to criteria
df1.query("Region == 'East'")
````

````python
# select rows according to criteria
df1.query("(Region == 'East') & (Units < 10)")
````

````python
# reindex for date
df1.set_index('Order_Date', inplace=True)
df1.head(5)
````

````python
# select by date range
df1.loc['2018-10-1':'2018-12-31']
````

````python
df1.loc['2019-02']
````

````python
# resampling by date
df1.resample('Y').sum(numeric_only=True) # group by year
````

````python
# resampling by date
df1.resample('M').sum(numeric_only=True) # Group by month
````

````python
# resampling by date
df1.resample('3M').sum(numeric_only=True) #group by 3 months
````

````python
# new variable with 2 month averages
bimonthly = df1.resample('2M').mean(numeric_only=True) 
bimonthly
````

````python
# group by column values ​​and calculate statistics
df1.groupby(['Region']).sum(numeric_only = True)
````

````python
# group by two columns values ​​and calculate statistics
df1.groupby(['Region','Rep']).sum(numeric_only = True)
````

````python
# group by 2 columns and apply various aggregate functions to the units column
df1.groupby(['Region','Rep']).Units.agg(['sum', 'min', 'max', 'mean'])
````

## Read a new file from git hub

````python
from pandas.plotting import scatter_matrix
# File url
url = 'https://github.com/E-man85/ML-PY/raw/main/03-data/winequality.xls'
# Read directly from GitHub with pandas
df1 = pd.read_excel(url, sheet_name=0, engine='xlrd')
# Show first 5 rows of dataframe
df1.head()
````

````python
# correlation between columns
matcorr = df1.corr(numeric_only = True)
matcorr
````

````python
# Correlation between two columns
df1[['alcohol', 'pH']].corr()
````

````python
# identification of NaN in the dataframe
df1.isnull().sum()
````

````python
# Isolate control region and convert to binary
control = df1.loc[df1['type'].isin(['white'])].copy()
control['quality'] = (control['quality'] >= 7).astype(int)
control['quality'].sum()
````

````python
# group by values ​​and calculate statistics
zonasel = df1.groupby(['type']).mean()
zonasel
````

#### PLOTTING ####

````python
# variable charts in Panda
control['alcohol'].plot() # x=indice, y=valor
````

````python
# Historam
control[['fixed_acidity','volatile_acidity']].plot.hist(stacked=False, bins=200) 
````

````python
 # scatter plot
control.plot.scatter(x='fixed_acidity', y='volatile_acidity')
````

````python
control.plot.scatter(x='fixed_acidity', 
                     y='volatile_acidity', 
                     c='alcohol', 
                     s=50, 
                     colormap='jet', 
                     grid='on')
````

````python
# Scatter Matrix in Pandas
scatter_matrix(control[['fixed_acidity','volatile_acidity','alcohol']], 
               alpha=0.2, 
               figsize=(10,10), 
               diagonal='hist')
````

````python
# Pandas integration with Seaborn
import seaborn as sns
sns.set(style='whitegrid') #whitegrid, dark, white, ticks
sns.pairplot(control, 
             vars=control[['fixed_acidity','volatile_acidity','alcohol']], 
             hue='quality',
             diag_kind='kde')
````
